```python
import pandas as pd
csv_file = "litecoin.csv"

#Read in File data and store into Pandas data frame
litecoin_df = pd.read_csv(csv_file)
litecoin_df.head(10)
```

```python
#Round decimals and change Date to datetime data type
litecoin_df = pd.DataFrame(litecoin_df.round(2))
litecoin_df['Date'] = pd.to_datetime(litecoin_df['Date'])

#Check data types to confirm that date = datetime & all others are floats
litecoin_df.dtypes
```

```python
#Drop NaNs
litecoin_df = litecoin_df.dropna()
```

```python
#Check that all NaNs were dropped
rows_with_nan = []
for index, row in litecoin_df.iterrows():
    is_nan_series = row.isnull()
    if is_nan_series.any():
        rows_with_nan.append(index)

print(rows_with_nan)
```

```python
#Set date as index

litecoin_df = litecoin_df.set_index('Date')
litecoin_df.head(20)
```

```python
#Introduce new dataframe of only the dates desired
new_litecoin_df = pd.DataFrame(litecoin_df.loc['2018-04-02':'2021-04-01'])
new_litecoin_df
```

```python
#Export to CSV
new_litecoin_df.to_csv('clean_litecoin.csv')
```
