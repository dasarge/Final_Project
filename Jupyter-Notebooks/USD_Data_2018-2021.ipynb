```python
import pandas as pd
csv_file = "dollar.csv"

#Read in File data and store into Pandas data frame
dollar_df = pd.read_csv(csv_file)
dollar_df.head(10)
```

```python
#Round decimals and change Date to datetime data type
dollar_df = pd.DataFrame(dollar_df.round(2))
dollar_df['Date'] = pd.to_datetime(dollar_df['Date'])

#Check data types to confirm that date = datetime & all others are floats
dollar_df.dtypes
```

```python
#Drop NaNs
dollar_df = dollar_df.dropna()
```

```python
#Check that all NaNs were dropped
rows_with_nan = []
for index, row in dollar_df.iterrows():
    is_nan_series = row.isnull()
    if is_nan_series.any():
        rows_with_nan.append(index)

print(rows_with_nan)
```

```python
#Set date as index

dollar_df = dollar_df.set_index('Date')
dollar_df.head(20)
```

```python
#Introduce new bitcoin dataframe of only the dates desired
new_dollar_df = pd.DataFrame(dollar_df.loc['2018-04-02':'2021-04-01'])
new_dollar_df
```

```python
#Export to CSV
new_dollar_df.to_csv('clean_dollar.csv')
```
