```python
import pandas as pd
csv_file = "S&P500.csv"

#Read in File data and store into Pandas data frame
SnP_df = pd.read_csv(csv_file)
SnP_df.head(10)
```

```python
#Round decimals and change Date to datetime data type
SnP_df = pd.DataFrame(SnP_df.round(2))
SnP_df['Date'] = pd.to_datetime(SnP_df['Date'])

#Check data types to confirm that date = datetime & all others are floats
SnP_df.dtypes
```

```python
#Drop NaNs
SnP_df = SnP_df.dropna()
```

```python
#Check that all NaNs were dropped
rows_with_nan = []
for index, row in SnP_df.iterrows():
    is_nan_series = row.isnull()
    if is_nan_series.any():
        rows_with_nan.append(index)

print(rows_with_nan)
```

```python
#Sort Date values then set date as index

SnP_df = SnP_df.sort_values(by='Date')
SnP_df = SnP_df.set_index('Date')
SnP_df.head(20)
```

```python
#Introduce new dataframe of only the dates desired
new_SnP_df = pd.DataFrame(SnP_df.loc['2018-04-01':'2021-04-01'])
new_SnP_df
```

```python
#Export to CSV
new_SnP_df.to_csv('clean_S&P500.csv')
```
