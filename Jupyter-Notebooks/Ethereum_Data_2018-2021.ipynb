```python
import pandas as pd
csv_file = "ethereum.csv"

#Read in File data and store into Pandas data frame
ethereum_df = pd.read_csv(csv_file)
ethereum_df.head(10)
```

```python
#Round decimals and change Date to datetime data type
ethereum_df = pd.DataFrame(ethereum_df.round(2))
ethereum_df['Date'] = pd.to_datetime(ethereum_df['Date'])

#Check data types to confirm that date = datetime & all others are floats
ethereum_df.dtypes
```

```python
#Drop NaNs
ethereum_df = ethereum_df.dropna()
```

```python
#Check that all NaNs were dropped
rows_with_nan = []
for index, row in ethereum_df.iterrows():
    is_nan_series = row.isnull()
    if is_nan_series.any():
        rows_with_nan.append(index)

print(rows_with_nan)
```

```python
#Set date as index

ethereum_df = ethereum_df.set_index('Date')
ethereum_df.head(20)
```

```python
#Introduce new dataframe of only the dates desired
new_ethereum_df = pd.DataFrame(ethereum_df.loc['2018-04-02':'2021-04-01'])
new_ethereum_df
```

```python
#Export to CSV
new_ethereum_df.to_csv('clean_ethereum.csv')
```
