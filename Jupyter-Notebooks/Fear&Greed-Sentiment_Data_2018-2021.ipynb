```python
import pandas as pd
csv_file = "Fear-and-Greed-Sentiment.csv"

#Read in File data and store into Pandas data frame
FnG_df = pd.read_csv(csv_file)
FnG_df.head(10)
```

```python
#Change Date to datetime data type, Index to Integer, Sentiment to string
FnG_df['Date'] = pd.to_datetime(FnG_df['Date'])
FnG_df['Fear and Greed Index'] = pd.to_numeric(FnG_df['Fear and Greed Index'])
FnG_df['Sentiment'] = FnG_df['Sentiment'].astype('string')

#Check data types to confirm that date = datetime & all others are floats
FnG_df.dtypes
```

```python
#Drop NaNs
FnG_df = FnG_df.dropna()
```

```python
#Check that all NaNs were dropped
rows_with_nan = []
for index, row in FnG_df.iterrows():
    is_nan_series = row.isnull()
    if is_nan_series.any():
        rows_with_nan.append(index)

print(rows_with_nan)
```

```python
#Sort Date values then set date as index

FnG_df = FnG_df.sort_values(by='Date')
FnG_df = FnG_df.set_index('Date')
FnG_df.head(20)
```

```python
#Introduce new dataframe of only the dates desired
FnG_df = pd.DataFrame(FnG_df.loc['2018-04-02':'2021-04-01'])
FnG_df
```

```python
#Export to CSV
FnG_df.to_csv('clean_Fear-and-Greed-Sentiment.csv')
```
